```python
#get urls for all images in github depository
#black_face_urls=[]
for i in range(1000,2000):
    url = 'https://github.com/ninadewi/CS231/raw/main/blackfaces2/'+str(i)+'.png'
    black_face_urls.append(url)
```

```python
import boto3, requests
session = boto3.Session(profile_name='default')
rekognition = session.client('rekognition')
#rekog_black_ages = []
#urls = ['s3://cs231/asianfaces/0.png','s3://cs231/asianfaces/1.png']

for i in range(1850,2000):
    response = requests.get(black_face_urls[i])
    print(black_face_urls[i])
    response_content = response.content
    rekognition_response = rekognition.detect_faces(Image={'Bytes': response_content}, Attributes=['ALL'])
    agerange=rekognition_response['FaceDetails'][0]['AgeRange']
    age=agerange['Low']+agerange['High']/2
    rekog_black_ages.append(age) 
print(rekog_black_ages)
```
